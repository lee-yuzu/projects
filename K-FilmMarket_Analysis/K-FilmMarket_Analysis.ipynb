#### [3] 데이터 전처리 <hr>

```python
# [3-1] 연도별 매출액 데이터프레임 전처리
# (1) 연도별 총 매출액 및 관객수
# 한국 데이터 슬라이싱
totalProfit_by_yearDF = totalProfit_by_yearDF.iloc[:-2,0:6]

# 결측치 채우기
totalProfit_by_yearDF.loc[0, '연도'] = '연도'

# 첫 행을 컬럼으로 지정            
header = totalProfit_by_yearDF.iloc[0]                     
totalProfit_by_yearDF = totalProfit_by_yearDF[1:]
totalProfit_by_yearDF.rename(columns=header, inplace=True)

 # int 형변환
totalProfit_by_yearDF = totalProfit_by_yearDF.astype({'연도':'int', '매출액':'int', '관객수':'int'})
totalProfit_by_yearDF.dtypes
```

```python
# (2) 지역별, 연도별 상영편수 및 매출액
# 데이터 프레임 합치기
region_DF = pd.concat([df_04,df_05,df_06,df_07,df_08,df_09,df_10,df_11,df_12,df_13,df_14,df_15,df_16, df_17, df_18, df_19, df_20, df_21, df_22, df_23, df_24], axis=0)

# 컬럼이름 변경
region_DF.rename(columns={'■ 지역별 점유율' : '지역',
                          'Unnamed: 1':'상영편수',
                          'Unnamed: 2':'매출액',
                          'Unnamed: 3':'관객수',
                          'Unnamed: 4':'점유율'}, inplace=True)

# 필요없는 행, 열 삭제
region_DF.drop(index=[0,1,3,4], columns=['Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12'], inplace=True)

# 내보내기 
# region_DF.to_excel('region_DF.xlsx')
region_DF
```

```python
# (3) 1인당 영화관람 횟수
# 행, 열 전환
watch_DF = watch_DF.transpose()

# 컬럼이름 변경
watch_DF.rename(columns={0:'전체',
                         1:'한국영화',
                         2:'외국영화',}, inplace=True)

# 필요없는 행, 열 삭제
watch_DF.drop(index='Unnamed: 0', inplace=True)
               
# 행, 열 전환
watch_DF
```

```python
# (4) 관람 가격
# 필요없는 행, 열 삭제
ticket_DF.drop(index=[0,1,2,3,4], inplace=True)

# 인덱스 재설정
ticket_DF.reset_index(drop=True, inplace=True)
ticket_DF
```

```python
# (5) 장르
# 필요없는 행 삭제
genre_DF.drop(index=[0,1,2,3,5], inplace=True)

# 첫 행을 컬럼으로 지정            
header = genre_DF.iloc[0]                     
genre_DF = genre_DF[1:]
genre_DF.rename(columns=header, inplace=True)

# 필요없는 열 삭제
genre_DF.drop(columns=['수입사','영화유형','국적','영화구분'], inplace=True)

# 결측치 채우기
genre_DF['장르'].fillna('성인물(에로)', inplace=True)
genre_DF

# 순번 컬럼 인덱스 설정
genre_DF.set_index('순번', inplace=True)

# 개봉일 컬럼에서 연도 컬럼 생성
genre_DF.개봉일 = genre_DF.개봉일.astype('datetime64[ns]') # datatime으로 형변환
genre_DF['연도'] = genre_DF.개봉일.dt.year

# 기간을 벗어난 1977, 2003년 행 제거
genre_DF = genre_DF[(genre_DF.연도 != 1977) & (genre_DF.연도 != 2003)]
```
